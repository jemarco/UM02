```python
import pandas as pd
```

```python
df = pd.read_csv('D:\\UTSEUS\\visualization\\Projet\\Interactive\\UM02_interactive_project\\top50.csv')
df = df[['Track.Name', 'Artist.Name', 'Genre', 'Popularity']]
df
```

```python
#make tables by genre
df = df.sort_values(by ='Genre')
df.reset_index(drop=True)
```

```python
df_copy = pd.DataFrame(columns=['Track.Name', 'Artist.Name', 'Genre', 'Popularity'])
#new_df = pd.DataFrame(fake_data).set_index('id')[3:].copy(deep=True)
```

```python
track_name = []
artist_name = []
genre = []
popularity = []
genrenb = []
genre_number = 0
musics_in_genre = []

while len(df)>1:
    c = 1
    last = 0
    bigger = 1
    genre_number = genre_number+1

    for i in range (1,len(df)):
        if df.iloc[i-1,2] == df.iloc[i,2]:
            c+= 1
        else : 
            c = 0
        if c >= bigger:
            bigger = c
            last = i

    first = last-bigger
    musics_in_genre.append(bigger)
    

    for j in range (first, last+1): #+1 because missing the last one
        track_name.append(df.iloc[j,0])
        artist_name.append(df.iloc[j,1])
        genre.append(df.iloc[j,2])
        popularity.append(df.iloc[j,3])
        genrenb.append(genre_number)

    df = df[df.Genre != str(genre[-1])]
    df.reset_index(drop=True)
```

```python
by_genre = {'name': track_name, 'artist': artist_name, 'genre': genre, 'popularity': popularity, 'genre number': genrenb}
df_by_genre = pd.DataFrame(by_genre)
df_by_genre
```

```python
df_by_genre.drop_duplicates(subset='name', inplace=True)
df_by_genre.reset_index(drop=True)
#trouble with the genre number of the "ransom" music. fixing it manually
```

```python
musics_in_genre.append(1)
for i in range (43,50):
    df_by_genre.iloc[i, 4] = df_by_genre.iloc[i, 4]+1
df_by_genre
```

```python
df_genre_only = df_by_genre.copy()
df_genre_only.drop_duplicates(subset='genre', inplace=True)
df_genre_only.reset_index(drop=True)
df_genre_only = df_genre_only[['genre', 'genre number']]
df_genre_only['musics in genre'] = musics_in_genre
df_genre_only.reset_index(drop=True)
```

```python
df_genre_only.to_csv('D:\\UTSEUS\\visualization\\Projet\\Interactive\\UM02_interactive_project\\top50_genre_only.csv')
```
