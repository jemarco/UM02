```python
df_credits = pd.read_csv('D:\\UTSEUS\\visualization\\Projet\\Network\\credits.csv')
df_credits = df_credits.dropna()
df_credits = df_credits[df_credits['cast'] != '[]']
df_credits.reset_index(drop=True)
```

```python
#make a list of each cast member, no duplicate
#first we want a list of the cast members only the name
import ast

for i  in range (len(df_credits)):
    actors = []
    a = df_credits.iloc[i,0]
    a = ast.literal_eval(a)
    for j in range (len(a)):
        try:
            actor = str(a[j])
            actor = actor.split("name': '")[1].strip()
            actor = actor.split("',")[0].strip()
            actors.append(actor)
        except:
            df_credits.drop(df_credits.index[i])
    df_credits.iloc[i, 0] = str(actors)

df_credits
```

```python
# We want a list with the following format : "[[actor1,[network1]],[actor2,[network2]]" 
#to make it a datafram that can be read by gephi

list_network = []

for i in range (len(list_actors)):
    list_network.append([list_actors[i],[]])
    for j in range (len(df)):
        c=0
        for k in range (len(df.iloc[j,0])):
            if list_actors[i] == df.iloc[j,0][k]:
                c+=1
        if c != 0:
            for l in range(len(df.iloc[j,0])):
                list_network[i][1].append(df.iloc[j,0][l])
    list_network[i][1].remove(list_network[i][0])
                
list_network          
                
    
```

```python
df_network = pd.DataFrame(list_network,columns=['actor','actor network'])
df_network
```
